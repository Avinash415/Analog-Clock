```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create figure and axes
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

# Plot data on the first subplot
ax1.plot(x, y1, label='Sin(x)')
ax1.set_ylabel('Sin(x)')
ax1.set_title('Multiple Figures and Axes Example')

# Plot data on the second subplot
ax2.plot(x, y2, label='Cos(x)')
ax2.set_xlabel('X')
ax2.set_ylabel('Cos(x)')

# Add text
ax1.text(5, 0.5, 'Text on Subplot 1', fontsize=12, color='blue')
ax2.text(5, 0.5, 'Text on Subplot 2', fontsize=12, color='red')

# Add grid
ax1.grid(True)
ax2.grid(True)

# Add legend
ax1.legend()
ax2.legend()

# Show the plot
plt.show()
```
